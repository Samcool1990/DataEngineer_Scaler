```python
import numpy as np
import yfinance as yf
import os
import time
```

```python
while(1==1):
	for crypt in ['BTC-USD','ETH-USD','USDT-USD','BNB-USD','USDC-USD']:
		data = yf.download(tickers=crypt, period = '2h', interval = '1m')
		data = data.tail(1)
		data['Type'] = crypt
		data = data[['Type','Close']]
		print(data.info())
		data = data.to_csv("./data/current_{}.csv".format(crypt.replace('-','_')), index=False, header = False)
		print(type(data))
		os.system("kafka-console-producer --broker-list localhost:9092 --topic cryptos < ./data/current_{}.csv".format(crypt.replace('-','_')))
	time.sleep(30)
```
